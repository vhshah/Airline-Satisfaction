```python
columns=[ 'Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class',
       'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'Dept_delay_more_20min', 
       'satisfaction']
df=df[columns]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
df[['Age','Flight Distance']]=ss.fit_transform(df[['Age','Flight Distance']])
```

```python
df.to_csv('Preprocessed.csv',index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],random_state=2)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
x_test
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```

```python
lr.coef_[0]
```

```python
summary_table=pd.DataFrame()
summary_table['Coefficients']=lr.coef_[0]
summary_table.index=x_train.columns
summary_table['Std_Dev']=np.std(df)
summary_table['Importance']=summary_table['Coefficients']*summary_table['Std_Dev']
```

```python
summary_table.sort_values(by='Importance',key=lambda x: abs(x),inplace=True,ascending=False)
```

```python
summary_table
```

```python
fig,ax=plt.subplots(figsize=(15,10))
plt.xticks(rotation=90)
plt.yticks(np.arange(-3,1,0.1))
sns.barplot(x=summary_table.index,y=summary_table.Importance)
plt.title('Logistic Regression Feature Importance')
plt.xlabel('Feature Name')
plt.savefig('LogisticRegression.png')
plt.show()
```

```python
import xgboost as xg
```

```python
summary_table_xg=pd.DataFrame()
```
